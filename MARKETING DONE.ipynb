```python
data1=data.drop(['Customer id'],axis=1)
data1
```

```python
data1=pd.get_dummies(data1,dtype=int)
data1.head()
```

```python
cor_mat=data1.corr()
cor_mat
```

```python
y=data1['Purchased']#copied to check
x=data1.drop('Purchased',axis=1)#removed for model
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.50,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression()
classifier.fit(x_train,y_train)#for fitting and training the model
```

```python
x
```

```python
y
```

```python
ypred=classifier.predict(x_test)
ypred
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,ypred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,ypred)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(ypred,y_test)
```

```python
Results=pd.DataFrame(columns=['Price','Predicted'])
Results['Price']=y_test
Results['Predicted']=ypred
Results=Results.reset_index()
Results['ID']=Results.index
Results.head(15)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.lineplot(x='ID',y='Price',data=Results.head(50))
sns.lineplot(x='ID',y='Predicted',data=Results.tail(50))
plt.plot()
```
